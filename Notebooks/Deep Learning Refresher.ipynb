### Vanishing Gradient Problem

Occurs when using activation funtions such as the sigmoid curve and during back propagation  - chain of derivatives the values become too small such that the old weights are almost equal to the new weights the impact of which means the weights never change - are not getting updated or updated with a smaller value. 

The solution is to use other activation functions:
- Tanh
- ReLU
- Leaky ReLU
- PReLU

## Activation Functions

The following are some of the activation functions:
- Sigmoid
- Tanh
- ReLu
- Leaky ReLU - solves the dead neuron problem
- ELU - Exponential linear Unit
- Softmax

Both Sigmoid and tanh lead to a vanishing gradient problem:

Solution: 

For Binary Classification
- In hidden layers use ReLU:If the convergence is not happening with ReLU you can change to ELU or PReLU. 
- In binary classification: Use Sigmoid activation function in the output layer

For multi class classification problem:
- Use softmax activation in the output layer
- In the hidden layer you can use ReLU and If the convergence is not happening with ReLU you can change to ELU or PReLU.

For Regression:
- In the hidden layers use ReLu or any variations of ReLU
- In the output layer use Linear activation functions: and in this case there will be a separate loss function.


## Loss Functions

**Regression**

In regression the type of losses:
- Mean Squared Error(MSE)
- Mean Absolute Error(MAE)
- Huber Loss

Loss function vs Cost Funtion: In loss function one record is passed while in cost function  a batch of records is passsed through the neural network. 

In MSE - has the advantage of 1 local or global minima also easily differentiable since it's a quadratic equation. we penalizing the error, hence this is more sensitive to outliers. Not robust to outliers.
in MAE - we are not penalizing the error by taking the absolute value - hence more robust to outliers. Time consuming compared to MSE.

Huber loss - is a combination of MAE & MSE

Root Mean Squrared Error(RMSE): MSE and RMSE both serve as important metrics for assessing the accuracy of regression models, with RMSE being more interpretable due to its unit consistency with the target variable.
